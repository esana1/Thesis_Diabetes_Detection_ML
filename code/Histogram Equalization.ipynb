```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
img = cv2.imread('iris.png', 0)

# Perform histogram equalization
equ = cv2.equalizeHist(img)

# Display original and equalized image
cv2.imshow('Original image', img)
cv2.imshow('Equalized Image', equ)

# Display original and equalized histogram
plt.figure(figsize=(10,5))
plt.subplot(121), plt.hist(img.ravel(),256,[0,256])
plt.title('Original Histogram')
plt.subplot(122), plt.hist(equ.ravel(),256,[0,256])
plt.title('Equalized Histogram')

plt.show()

# Save the equalized image
cv2.imwrite('output.jpg', equ)

# Close the windows when any key is pressed
cv2.waitKey(0)
cv2.destroyAllWindows()
```
